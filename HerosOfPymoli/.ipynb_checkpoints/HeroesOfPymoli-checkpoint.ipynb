```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load 
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#get unique values for column Screen Name to determine num of unique players
players_df= purchase_data["SN"].unique()
total_players = len(players_df)
#create a summary data frame 
total_players_df=pd.DataFrame({"Total Players": [total_players]})
total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Get unique values for column item names to determine num of unique items
items_list = purchase_data["Item Name"].unique()
```

```python
#Create and pass values to a summary data frame
cleaned_purchase=pd.DataFrame({"Number of Unique Items":[len(items_list)],
                              "Number of Purchases":[len(purchase_data)],
                              "Average Price":[purchase_data["Price"].mean()],
                              "Total Revenue": [purchase_data["Price"].sum()]})
cleaned_purchase.style.format({'Total Revenue': "${0:,.2f}", 'Average Price':"${0:,.2f}" })
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Count male,female and other players
p = purchase_data["Gender"].value_counts()
#Calc percentage of players by gender
p_percent=p/len(purchase_data)*100
#pass values to Data Frame for display and formatting
gender_analysis=pd.DataFrame({"Total Count":p,"Percentage of Players":p_percent})
gender_analysis.style.format({'Percentage of Players': "{0:,.2f}%"})
```


## Purchasing Analysis (Gender)

```python
#Basic calc to find count,avg price, avg per person spent by gender
g_count=purchase_data.groupby(['Gender']).count()['SN'].rename("Total Purchase Count")
g_sum=purchase_data.groupby(['Gender']).sum()['Price'].rename("Total Purchase Value")
g_avg=purchase_data.groupby(['Gender']).mean()['Price'].rename("Average Purchase Value")
#Average purchase price per person
h= purchase_data.groupby(['Gender','SN']).agg({'Price':'sum'})
Avgpp=h.groupby('Gender').mean()
#pass values to dataframe for display
purchase_analysis=pd.DataFrame({"Purchase Count": g_count,"Total Purchase Value":g_sum,"Average Purchase Value":g_avg})

```

```python
#update to data frame with the avg purchase price per person values
purchase_analysis['Avg Total Purchase per Person']=Avgpp
purchase_analysis.style.format({'Total Purchase Value':"${0:,.2f}","Average Purchase Value":"${0:,.2f}","Avg Total Purchase per Person":"${0:,.2f}"})
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#create age bins and label the bins
bins=[0,9,14,19,24,29,34,39,40]

group_names=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
#slice data into the above bins
purchase_data["Age Bracket"]=pd.cut(purchase_data["Age"],bins,labels=group_names)
```

```python
#Basic calc on the binned data to find count and percent of players by age bracket
bin_data = purchase_data.groupby("Age Bracket").agg({"Purchase ID":"count"})
bin_data['Percentage of Players']=bin_data["Purchase ID"]*100/len(purchase_data)
bin_data.style.format({'Percentage of Players': "{0:,.2f}%"})
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_bin1 =  purchase_data.groupby("Age Bracket").count()["Purchase ID"]
purchase_bin2 =  purchase_data.groupby("Age Bracket").mean()["Price"]
purchase_bin3 =  purchase_data.groupby("Age Bracket").sum()["Price"]
#pass values to data frame for display
purchase_bin=pd.DataFrame({"Purchase Count":purchase_bin1,"Avg Purchase Price":purchase_bin2,"Total Purchase":purchase_bin3})
```

```python
#calc avg purchase total per person

i=purchase_data.groupby(['Age Bracket','SN']).agg({"Price":"sum"})
avppa=i.groupby(['Age Bracket']).mean()
#add the new column to existing data frame
purchase_bin['Average Total per Person']=avppa
purchase_bin.style.format({'Average Purchase Price': "${0:,.2f}",'Total Purchase': "${0:,.2f}","Average Total per Person":"${0:,.2f}"})
```

## Top Spenders

```python
#Badic calc to find count, avg price and total purchase value by person(Screen name)
ts_count= purchase_data.groupby("SN").count()["Purchase ID"]
ts_avg =purchase_data.groupby("SN").mean()["Price"]
ts_sum=purchase_data.groupby("SN").sum()["Price"]
#Pass values to data frame for display
ts = pd.DataFrame({"Purchase Count":ts_count,"Average Purchase Price":ts_avg,"Total Purchase Value":ts_sum})
ts.style.format({'Average Purchase Price': "${0:,.2f}",'Total Purchase Value': "${0:,.2f}"})
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Calculate price, count,total purchase value by item id and item name
popular_items=purchase_data.groupby(["Item ID","Item Name","Price"]).agg({"Purchase ID":"count","Price":{"Total Purchase Value":"sum"}})
#format the values to show $ for currency
popular_items.columns=popular_items.columns.droplevel(level=0)
popular_items.style.format({'Price': "${0:,.2f}",'Total Purchase Value': "${0:,.2f}"})
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
#Sort the above items in descending order to show the most profitable items

popular_sorted =popular_items.sort_values("Total Purchase Value", ascending=False)
popular_sorted.style.format({'Price': "${0:,.2f}",'Total Purchase Value': "${0:,.2f}"})
```
